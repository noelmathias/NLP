```python
sentence = "On Tuesday, Elon Musk announced that Tesla will open a new manufacturing facility in Berlin  by January 2027 , investing over $1 billion  into the project."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
```
