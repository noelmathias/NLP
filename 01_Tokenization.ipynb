```python
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt')
```

```python
corpus = """Hello! Welcome to my place.
Please do come again soon! we'll be waiting.
"""
```

```python
print(corpus)
```

```python
#Tokenization
#paragraph --> sentences
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
for sentence in documents:
    print(sentence)
```

```python
#sentences --> Words
from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
```

```python
for word in words:
    print(word)
    
```

```python
#splitting punctuations
from nltk.tokenize import wordpunct_tokenize
word_punct = wordpunct_tokenize(corpus)
```
